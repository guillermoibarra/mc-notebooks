```python
import random
import math
import matplotlib.pyplot as plt

# Number of random points
N = 100000
inside_x = []
inside_y = []
outside_x = []
outside_y = []
M = 0

for _ in range(N):
    x, y = random.uniform(0,1), random.uniform(0,1)
    if x**2 + y**2 <= 1:
        M += 1
        inside_x.append(x)
        inside_y.append(y)
    else:
        outside_x.append(x)
        outside_y.append(y)

# Estimate of pi
pi_estimate = 4 * M / N
print(f"Estimated value of pi: {pi_estimate}")

# Plotting
fig, ax = plt.subplots()
# Square boundary
square = plt.Rectangle((0,0),1,1, fill=False)
ax.add_patch(square)
# Quarter-circle boundary
theta = [i * 0.01 for i in range(0, 314)]
circle_x = [round(math.cos(t)*1, 4) for t in theta]
circle_y = [round(math.sin(t)*1, 4) for t in theta]
ax.plot(circle_x, circle_y, color='green')

# Plot points
ax.scatter(inside_x, inside_y, color='blue', s=1, label='Inside Quarter-Circle')
ax.scatter(outside_x, outside_y, color='red', s=1, label='Outside Quarter-Circle')

plt.title(f"Monte Carlo Simulation to Estimate \nEstimated  = {pi_estimate}")
plt.legend()
plt.show()
```
