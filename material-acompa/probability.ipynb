# Eventos Aleatorios

Ejemplo: Simulación de lanzamientos de una moneda.

- Descripción: Simular 100 lanzamientos de una moneda justa y contar cuántas veces sale aguila y sol.
- Python: Utilizar la librería `random` para generar resultados aleatorios y visualizar la distribución con `matplotlib`.

```python
import random
import matplotlib.pyplot as plt

# Simulación de 100 lanzamientos
lanzamientos = [random.choice(['Aguila', 'Sol']) for _ in range(10000)]
aguila = lanzamientos.count('Aguila')
sol = lanzamientos.count('Sol')

# Visualización
plt.bar(['Aguila', 'Sol'], [aguila, sol])
plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.title('Simulación de Lanzamientos de una Moneda')
plt.show()
```

### 2
La probabilidad de al menos un "Cara" es 1 menos la probabilidad de obtener cero "Caras".

$$𝑃(Al menos un Cara) = 1 − 𝑃(Cero Caras)=1−\frac{1}{2}^{10}= 1 − \frac{1}{1024} ≈ 0.999$$

```python
from scipy.stats import binom

n = 10
p = 0.5
k = 0

prob_cero_caras = binom.pmf(k, n, p)
prob_al_menos_un_cara = 1 - prob_cero_caras

print(f'Probabilidad de al menos un "Cara": {prob_al_menos_un_cara:.4f}')
```

```python
# 3
import random

N = 1000
exactamente_una_cara = 0

for _ in range(N):
    lanzamientos = [random.choice(['Cara', 'Cruz']) for _ in range(2)]
    if lanzamientos.count('Cara') == 1:
        exactamente_una_cara += 1

probabilidad = exactamente_una_cara / N
print(f'Probabilidad simulada de obtener exactamente una "Cara": {probabilidad:.4f}')
```

```python
# 4
from itertools import product

lanzamientos_posibles = list(product(['Cara', 'Cruz'], repeat=3))
favorables = [seq for seq in lanzamientos_posibles if ('Cara', 'Cara') in zip(seq, seq[1:])]

probabilidad = len(favorables) / len(lanzamientos_posibles)
print(f'Probabilidad de obtener dos "Caras" consecutivas en tres lanzamientos: {probabilidad:.3f}')
```

# Distribución de Poisson (The Poisson Distribution)

- Ejemplo: Modelar el número de llamadas que recibe un centro de atención en una hora.
- Python: Utilizar `scipy.stats.poisson` para analizar la distribución

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Parámetros
mu = 50  # Tasa promedio de llamadas por hora

# Distribución de Poisson
k = np.arange(0, 100)
probabilidades = poisson.pmf(k, mu)

# Visualización
plt.stem(k, probabilidades)
plt.xlabel('Número de llamadas')
plt.ylabel('Probabilidad')
plt.title('Distribución de Poisson')
plt.show()
```

### 5

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, binom
La distribución de Poisson es una aproximación de la binomial cuando $𝑛$  es grande y $p$ es pequeño $(\lambda = 𝑛 \cdot p)$. La distribución geométrica modela el número de intentos hasta el primer éxito, mientras que Poisson modela el número de eventos en un intervalo fijo.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, binom

# Parámetros
n_binomial = 1000
p_binomial = 0.004  # lambda = n * p = 4

# Distribuciones
k_poisson = np.arange(0, 15)
prob_poisson = poisson.pmf(k_poisson, mu=4)  # Corrected parameter name

prob_binomial = binom.pmf(k_poisson, n_binomial, p_binomial)

# Visualización
plt.stem(k_poisson, prob_poisson, linefmt='b-', markerfmt='bo', basefmt=" ", label='Poisson')
plt.stem(k_poisson, prob_binomial, linefmt='r--', markerfmt='ro', basefmt=" ", label='Binomial (n=1000, p=0.004)')
plt.xlabel('Número de Eventos')
plt.ylabel('Probabilidad')
plt.title('Comparación entre Poisson y Binomial')
plt.legend()
plt.show()
```
