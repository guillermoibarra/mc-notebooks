```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Needed for 3D plotting

# Set up a figure with three subplots side by side
fig = plt.figure(figsize=(18, 5))

##############################################################################
# 1) 1D Integral Visualization
##############################################################################
ax1 = fig.add_subplot(1, 3, 1)

# Define the 1D function f(x) = x^2 on [0, 1]
x = np.linspace(0, 1, 200)
y = x**2

# Plot f(x) and fill under the curve to illustrate the "area under the curve"
ax1.plot(x, y, 'b', label=r'$f(x) = x^2$')
ax1.fill_between(x, y, color='blue', alpha=0.3)
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.set_title('1D Integral')
ax1.legend()

##############################################################################
# 2) 2D Integral Visualization
##############################################################################
ax2 = fig.add_subplot(1, 3, 2, projection='3d')

# Define the 2D function f(x,y) = x^2 + y^2 on [-1, 1] x [-1, 1]
X = np.linspace(-1, 1, 50)
Y = np.linspace(-1, 1, 50)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2

# Create a surface plot
surf = ax2.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.8)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('f(x,y)')
ax2.set_title('2D Integral')

# Add a color bar to show the mapping from Z values to colors
fig.colorbar(surf, ax=ax2, shrink=0.5, aspect=10)

##############################################################################
# 3) 3D Integral Visualization
##############################################################################
ax3 = fig.add_subplot(1, 3, 3, projection='3d')

# Define the 3D function f(x,y,z) = x^2 + y^2 + z^2 on [-1, 1]^3
N = 200  # Number of sample points for the scatter
x3 = np.random.uniform(-1, 1, N)
y3 = np.random.uniform(-1, 1, N)
z3 = np.random.uniform(-1, 1, N)

# Evaluate the function
values = x3**2 + y3**2 + z3**2

# Scatter plot in 3D, coloring each point by its function value
p = ax3.scatter(x3, y3, z3, c=values, cmap='plasma', alpha=0.8)
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_zlabel('z')
ax3.set_title('3D Integral')

# Add a color bar to interpret function values
fig.colorbar(p, ax=ax3, shrink=0.5, aspect=10)

plt.suptitle('Visualizing 1D, 2D, and 3D Functions for Integration')
plt.show()
```
