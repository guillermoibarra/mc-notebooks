# Ejercicios para Despertar

1. Inversión de Cadenas y Palíndromos
   - Pide al usuario que ingrese una cadena de texto.
   - Escribe una función que invierta esa cadena.
   - Verifica si la cadena original es un palíndromo (se lee igual de izquierda a derecha y de derecha a izquierda).
   - Imprime la cadena invertida y un mensaje indicando si es un palíndromo o no.
2. FizzBuzz
   - Imprime los números del 1 al 50.
   - Para múltiplos de 3, imprime "Fizz" en lugar del número.
   - Para múltiplos de 5, imprime "Buzz".
   - Para múltiplos de ambos (3 y 5), imprime "FizzBuzz".
3. Calculadora Sencilla
   - Pregunta al usuario por dos números y un operador (por ejemplo, +, -, *, /).
   - Realiza la operación solicitada.
   - Muestra el resultado.
   - Continúa pidiendo más operaciones hasta que el usuario decida salir (por ejemplo, ingresando "salir").
4. Juego de Adivinar el Número
   - El programa elige un número aleatorio entre 1 y 50
   - El usuario intenta adivinar el número.
   - El programa indica si la suposición es "Muy alta" o "Muy baja" hasta que el usuario acierte.
   - Lleva un conteo de los intentos.
5. Operaciones con Listas.
   - Pide al usuario que ingrese varios números (por ejemplo, 5).
   - Almacena esos números en una lista.
   - Imprime:
       - La suma de los números.
       - El promedio.
       - El valor máximo y mínimo.
   - Ordena la lista y muéstrala por pantalla.

```python
# Solución propuesta no. 1

# Pedir al usuario una cadena de texto
texto = input("Escribe una cadena de texto: ")

# Invertir la cadena (Python permite usar slicing [::-1])
texto_invertido = texto[::-1]

# Verificar si es palíndromo (ignorando mayúsculas/minúsculas)
es_palindromo = (texto.lower() == texto_invertido.lower())

# Mostrar resultados
print(f"La cadena invertida es: {texto_invertido}")
print(f"¿Es palíndromo?: {es_palindromo}")
```

```python
# Solución propuesta no 2

for numero in range(1, 51):
    if numero % 3 == 0 and numero % 5 == 0:
        print("FizzBuzz")
    elif numero % 3 == 0:
        print("Fizz")
    elif numero % 5 == 0:
        print("Buzz")
    else:
        print(numero)
```

```python
# Solución propuesta no 3

while True:
    # Pedir el primer número
    num1_str = input("Ingresa el primer número (o 'salir' para terminar): ")
    
    # Revisar si el usuario quiere salir
    if num1_str.lower() == "salir":
        print("¡Calculadora finalizada!")
        break
    
    # Pedir el segundo número
    num2_str = input("Ingresa el segundo número: ")
    
    # Pedir el operador
    operador = input("Ingresa el operador (+, -, *, /): ")
    
    # Convertir las entradas a números de punto flotante
    try:
        num1 = float(num1_str)
        num2 = float(num2_str)
    except ValueError:
        print("Error: Por favor ingresa valores numéricos válidos.")
        continue
    
    # Realizar la operación
    if operador == '+':
        resultado = num1 + num2
    elif operador == '-':
        resultado = num1 - num2
    elif operador == '*':
        resultado = num1 * num2
    elif operador == '/':
        if num2 == 0:
            print("Error: No se puede dividir entre cero.")
            continue
        else:
            resultado = num1 / num2
    else:
        print("Operador inválido. Intenta de nuevo.")
        continue
    
    print(f"El resultado de {num1} {operador} {num2} es: {resultado}")
```

```python
# Solución propuesta 4

import random

# Generar un número aleatorio entre 1 y 50
numero_secreto = random.randint(1, 50)
intentos = 0

while True:
    respuesta = input("Adivina el número (entre 1 y 50) o escribe 'salir' para terminar: ")
    
    # Permitir al usuario salir manualmente
    if respuesta.lower() == 'salir':
        print("¡Gracias por jugar!")
        break
    
    # Verificar que el input sea un número válido
    try:
        numero_adivinado = int(respuesta)
    except ValueError:
        print("Por favor, ingresa un número entero válido.")
        continue
    
    intentos += 1
    
    if numero_adivinado < numero_secreto:
        print("Muy bajo. Intenta de nuevo.")
    elif numero_adivinado > numero_secreto:
        print("Muy alto. Intenta de nuevo.")
    else:
        print(f"¡Felicidades! Adivinaste el número en {intentos} intentos.")
        break
```

```python
# Solución 5

# Solicitar 5 números en una sola línea, separados por espacios
numeros_str = input("Ingresa 5 números separados por espacio: ")

# Dividir la cadena en elementos individuales
numeros_lista = numeros_str.split()

# Verificar que sean 5 elementos
if len(numeros_lista) != 5:
    print("No has ingresado exactamente 5 valores.")
else:
    # Intentar convertir cada elemento a float (o int)
    try:
        numeros = [float(x) for x in numeros_lista]
    except ValueError:
        print("Error: Asegúrate de ingresar solo números.")
        numeros = []
    
    # Si la conversión fue exitosa y tenemos 5 números
    if len(numeros) == 5:
        suma_numeros = sum(numeros)
        promedio = suma_numeros / len(numeros)
        maximo = max(numeros)
        minimo = min(numeros)
        numeros_ordenados = sorted(numeros)
        
        print(f"Suma: {suma_numeros}")
        print(f"Promedio: {promedio}")
        print(f"Máximo: {maximo}")
        print(f"Mínimo: {minimo}")
        print(f"Números ordenados: {numeros_ordenados}")
```
